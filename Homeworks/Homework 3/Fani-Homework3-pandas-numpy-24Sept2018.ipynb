Peform the following array operations using numpy

### 3.3.1 addition a+b 

```python
a= np.array([1,2,3,4],dtype=float)
b=np.array([5,6,7,8],dtype=float)
c=np.add(a,b)
print("Addition using function", c)
d=a+b
print("Addition using operation",d)
```

### 3.3.2 subtraction a-b

```python
a= np.array([1,2,3,4],dtype=float)
b=np.array([5,6,7,8],dtype=float)
c=np.subtract(a,b)
print("Subtraction using function", c)
d=a-b
print("Subtraction using operation",d)
```

### 3.3.3 multiplication a*b

```python
a= np.array([1,2,3,4],dtype=float)
b=np.array([5,6,7,8],dtype=float)
c=np.multiply(a,b)
print("Multiplication using function", c)
d=a*b
print("Multiplication using operation",d)
```

### 3.3.4 divsion a/b

```python
a= np.array([1,2,3,4],dtype=float)
b=np.array([5,6,7,8],dtype=float)
c=np.divide(b,a)
print("Division using function", c)
d=b/a
print("Division using operation",d)
```

### 3.3.5 modulo  a%b

```python
a= np.array([1,2,3,4],dtype=float)
b=np.array([5,6,7,8],dtype=float)
c=np.mod(a,b)
print("modulo using function", c)
d=a%b
print("modulo using operation",d)
```

### 3.3.6 power  a^b

```python
a= np.array([1,2,3,4],dtype=float)
b=np.array([5,6,7,8],dtype=float)
c=np.power(a,b)
print("power using function", c)
d=a**b
print("power using operation",d)
```

## Homework 3.5
Provide an interesting analysis of the data columns ( frequency or averages )

```python
my_dataset["score"].mean()
```

```python
#Number of rows in each score_phrase, we can see most of the games got 'Great' and 'Good' reviews
score_p={}
label=np.unique(my_dataset['score_phrase'])
#print(label)
for l in range(len(label)):
    score_p[label[l]]=0
for i in range(len(my_dataset)):
    score_p[my_dataset['score_phrase'][i]]+=1
    
print(score_p)
```

```python
#As we saw above, mean is around 7, we can use this to find all games that got above average score
score_filter = my_dataset["score"] > 7
score_filter
```

```python
#we can use above as a filter to display records that got score > 7
filter_reviews=my_dataset[score_filter]
filter_reviews
```

- Now that we saw how to make filters, we can create plots to observe the review distribution for the Xbox vs Playstation 4.

- We can make histogram for each console using pandas.DataFrame.plot method. This method utilizes matplotlib, the popular python plotting library 

```python
#call matplotlib to set up plotting inside a Jupyter notebook
%matplotlib inline
#Filter reviews to only have data about the Xbox one
my_dataset[my_dataset["platform"] == "Xbox One"]["score"].plot(kind="hist")
```

```python
#Filter reviews to only have data about the PS4
my_dataset[my_dataset["platform"] == "PlayStation 4"]["score"].plot(kind="hist")
```
