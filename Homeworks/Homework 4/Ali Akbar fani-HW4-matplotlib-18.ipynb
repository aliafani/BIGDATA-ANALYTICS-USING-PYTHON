#### Homework 4.1

4.1.a Create a figure with two subplots in a row.

Label the figures.

```python
t1 = np.arange(0.0,2*np.pi,0.05 )  

fig=plt.figure()
#for creating subplots the first value is height second is width and third is subplot
ax1=fig.add_subplot(221)
ax2=fig.add_subplot(222)

s = np.sin(t1)  

ax1.plot(t1, s)  
 
ax1.set_title('Figure 1 Sine of x',fontsize=9)
t2=np.arange(0.0,2*np.pi,0.001)  
tan_x=np.tan(t2)
ax2.plot(t2,tan_x)
ax2.set_title('Figure 2 Tangent of x',fontsize=9)
fig.suptitle('Figure with multiple subplots', fontsize=12)
plt.show()  
```

## Homework 4.2
 

Make a combined plot of a  red spiral with a scatter plot.

The red spiral has r from 0 to 3 with steps of 0.1 and
a theta of 2 * pi * r.

The scatter uses the same theta as the spiral but the radius r is offset by a random value ranging from -.5 to .5.
The size of the plotted locations is 100 * r^2 and the color uses theta.

Use an alpha value of .4 for the colors in the scatter plot.
Limit the size of the plot (rmax) to 2.5
 
 Should look something like:
<img src="./images/lineplot_with_scatterplot.png" width="300" height="200">


```python
import random

# Compute areas and colors

r_spiral = np.arange(0, 3.0, 0.01)
theta_spiral = 2 * np.pi * r_spiral

r_scatter=np.random.uniform(-0.5, 0.5,r_spiral.size)
r1=r_spiral+r_scatter
area = 100 * r_spiral**2
theta_scatter=2 * np.pi * np.random.uniform(low=-0.5, high=0.5, size=(150,))
colors = theta_spiral
fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')
ax.set_title("A line plot on polar axis")
ax.set_rlim(0,2.5)
ax.plot(0.5*theta_spiral, r_spiral, color='red')
ax.scatter(theta_spiral, r1, c=colors,cmap='jet',marker='o', s=area,alpha=0.4)
```

```python
from matplotlib.colors import LogNorm
n = 100000
x = np.random.normal(size=100000)
y = np.random.geometric(p=0.01, size=100000)
h=plt.hist2d(x,y,bins=100,norm=LogNorm(),cmap='jet')
plt.colorbar(h[3])
plt.title('2d Color Histogram')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax=fig.gca(projection='3d')
ax.set_title("3d Plot of 2d Histogram")

x = np.random.normal(size=100000)
y = np.random.geometric(p=0.01, size=100000)
hist, xedges, yedges = np.histogram2d(x, y, bins=100)
x,y=np.meshgrid(xedges[:-1],yedges[:-1])
ax.plot_surface(x,y,hist,rstride=1,cstride=1,cmap=plt.cm.jet)
plt.show()
```

# Stackplot 
- Stackplot are generated by plotting different datasets vertically on top of one another rather than overlapping with one   another.

```python
#Stackplots here we have example where we have list of days, sleeping, eating, working and playing
days=[1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing= [8,5,7,8,13]
#for legend we can give each line respective color and label as shown below
plt.plot([],[],color='m',label='Sleeping',linewidth=5)
plt.plot([],[],color='c',label='Eating',linewidth=5)
plt.plot([],[],color='r',label='Working',linewidth=5)
plt.plot([],[],color='k',label='Playing',linewidth=5)
plt.stackplot(days,sleeping,eating,working,playing, colors=['m','c','r','k'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Intresting graph')
plt.legend()
plt.show()
```

# Pie Plot
    Using matplotlib we can also draw Pie chart and make it more informative and attractive using great in-built feature

```python
#this is simple Pie plot where I have mentioned slices list and corresponding activities
slices=[7,2,3,13]
activites=['sleeping','eating','working','playing']
#specified color to each and every activities
cols=['c','m','r','brown']
plt.pie(slices,labels=activites,colors=cols)
plt.show()
```

```python
# we can make above pie chart more intresting by setting the starting angle 90 and highlighting specific portion of pie
slices=[7,2,3,13]
activites=['sleeping','eating','working','playing']
cols=['c','m','r','brown']
# here startangle and explode function will do the trick,here also to specify the value of each portion autopct is being used
plt.pie(slices,labels=activites,colors=cols,startangle=90,shadow=True,explode=(0,0.1,0,0),autopct='%1.1f')
plt.show()
```
